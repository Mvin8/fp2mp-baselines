# Evaluation

Evaluating `FP2MP-Eval` for every problem/solution pair

```python
from fp2mp_eval.utils import read_json, write_json

results = read_json('data/results.json')
```

```python
from tqdm import tqdm
from fp2mp_eval import FP2MPEval

N_JUDGES = 5

fp2mp = FP2MPEval(model='openai/gpt-4.1', n_judges=N_JUDGES)

for result in tqdm(results):
    problem = result['problem']
    solution = result['solution']
    evals = fp2mp.evaluate_case((problem, solution), max_workers=N_JUDGES)
    result['evaluations'] = [e.model_dump() for e in evals]
```
