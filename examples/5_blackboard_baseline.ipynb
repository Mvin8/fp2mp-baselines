````python
from langchain_openai import ChatOpenAI
from fp2mp_baselines import BlackboardBaseline
from fp2mp_baselines.config import config

def init_llm(model: str, temperature: float = 0.5):
    return ChatOpenAI(
        model=model,
        base_url=config.base_url,
        api_key=config.api_key,
        temperature=temperature,
    )

task = "Город Санкт-Петербург. Разработай план уплотнения городского центра."
````

````python
llm = init_llm("deepseek/deepseek-v4-flash")
baseline = BlackboardBaseline(llm=llm, iterations=3)
state = baseline.invoke_state(task)

````

````python
from IPython.display import Markdown, display

def show_blackboard_log(state):
    lines = ["# Blackboard log", ""]
    for i, entry in enumerate(state["log"], start=1):
        role = entry["metadata"].get("role_name", entry["agent"])
        lines.append(f"## {i}. {role}")
        lines.append(f"- agent_id: `{entry['agent']}`")
        lines.append(f"- event: `{entry['event']}`")
        lines.append("")
        lines.append("```python")
        lines.append(str(entry["response"]))
        lines.append("```")
        lines.append("")
    display(Markdown("\n".join(lines)))

show_blackboard_log(state)
````
